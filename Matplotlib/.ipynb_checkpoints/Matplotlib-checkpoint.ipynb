```python
import numpy as np
x= np.linspace(0,5,11)
y=x**2
```

functional

```python
plt.plot(x,y,'g')
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('X vs Y')
```

```python
#multiplot on same canvas
plt.subplot(1,2,1)
plt.plot(x,y,'g')
plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
#object oriented method
fig=plt.figure()
axes=fig.add_axes([0.1,0.1,0.8,0.8])#left bottom width height
axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Title')
```

```python
fig=plt.figure()#empty canvas
axes1=fig.add_axes([0.1,0.1,0.8,0.8])
axes2=fig.add_axes([0.2,0.5,0.4,0.3])
axes1.plot(x,y)
axes2.plot(y,x)
axes1.set_title('Larger plot')
axes2.set_title('Smaller plot')
```

```python
fig,axes=plt.subplots(nrows=1,ncols=2)
for current_ax in axes:
    current_ax.plot(x,y)
```

```python
fig,axes=plt.subplots(nrows=1,ncols=2)
axes[0].plot(x,y)
axes[1].plot(y,x)
axes[0].set_title('First Graph')
axes[1].set_title('Second Graph')
plt.tight_layout() #remove overlapping of subplots
```

```python
fig=plt.figure(figsize=(8,2))#width and height and dpi=pixel per inch
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
ax.set_title('Title')
ax.set_xlabel('X')
ax.set_ylabel('Y')
```

```python
fig,axes=plt.subplots(nrows=2,ncols=1,figsize=(8,2))#width and height and dpi=pixel per inch
axes[0].plot(x,y)
axes[1].plot(y,x)
plt.tight_layout()
```

```python
product_id=[1,2,3,4,5,6,7,8]
online_price=[123,567,343,123,170,388,400,250]
offline_price=[230,470,350,180,220,400,350,280]
chorbazar_price=[100,200,150,50,70,200,110,180]
```

```python
plt.plot(product_id,online_price,color='green',label='online')
plt.plot(product_id,offline_price,color='purple',label='offline')
plt.plot(product_id,chorbazar_price,color='blue',label='chorbazar')
plt.xlabel("Product")
plt.ylabel("Price")
plt.title("Product vs Price")
plt.legend()
```

```python
year=[2012,2013,2014,2015,2016,2017,2018,2019,2020,2021,2022,2023]
prices=[100,200,300,450,70,40,80,150,250,400,450,500]
plt.bar(year,prices,color='purple')
plt.xlabel('year')
plt.ylabel('prices')
plt.title('year by year price comparision')
plt.show()
```

```python
import pandas as pd
import random

# Create a list of 50 batsmen names
batsmen_names = ['Batsman ' + str(i) for i in range(1, 51)]

# Generate random data for strike rates and average runs
strike_rates = [round(random.uniform(50, 150), 2) for _ in range(50)]
average_runs = [random.randint(10, 100) for _ in range(50)]

# Create the DataFrame
batsman_data = pd.DataFrame({
    'Name': batsmen_names,
    'StrikeRate': strike_rates,
    'AverageRuns': average_runs
})
df = pd.DataFrame(batsman_data)

df
```

```python
plt.scatter(df['AverageRuns'],df['StrikeRate'])
plt.ylabel('strike rate')
plt.xlabel('average')
```
