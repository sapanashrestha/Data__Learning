```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline 
#view plot inside jupytern notebook
```

```python
import numpy as np
x= np.linspace(0,5,11)
y=x**2
```

```python
x
```

```python
y
```

functional

```python
plt.plot(x,y,'g')
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('X vs Y')
```

```python
#multiplot on same canvas
plt.subplot(1,2,1)
plt.plot(x,y,'g')
plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
#object oriented method
fig=plt.figure()
axes=fig.add_axes([0.1,0.1,0.8,0.8])#left bottom width height
axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Title')
```

```python
fig=plt.figure()#empty canvas
axes1=fig.add_axes([0.1,0.1,0.8,0.8])
axes2=fig.add_axes([0.2,0.5,0.4,0.3])
axes1.plot(x,y)
axes2.plot(y,x)
axes1.set_title('Larger plot')
axes2.set_title('Smaller plot')
```

```python
fig,axes=plt.subplots(nrows=1,ncols=2)
for current_ax in axes:
    current_ax.plot(x,y)
```

```python
fig,axes=plt.subplots(nrows=1,ncols=2)
axes[0].plot(x,y)
axes[1].plot(y,x)
axes[0].set_title('First Graph')
axes[1].set_title('Second Graph')
plt.tight_layout() #remove overlapping of subplots
```

```python
fig=plt.figure(figsize=(8,2))#width and height and dpi=pixel per inch
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
ax.set_title('Title')
ax.set_xlabel('X')
ax.set_ylabel('Y')
```

```python
fig,axes=plt.subplots(nrows=2,ncols=1,figsize=(8,2))#width and height and dpi=pixel per inch
axes[0].plot(x,y)
axes[1].plot(y,x)
plt.tight_layout()
```

```python
fig=plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2,label='X square')
ax.plot(x,x**3, label='X cube') 
ax.legend()
```

```python
#setting colors
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y,color='#FF8C00',linewidth=2,linestyle='-',marker='.')#rgb hex code for custom color
```
